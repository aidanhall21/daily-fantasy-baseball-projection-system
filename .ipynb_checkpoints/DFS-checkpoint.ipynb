```python
def draft_king_batters(single, double, triple, HR, RBI, R, BB, HBP, SB):
    
    return (3 * single) + (5 * double) + (8 * triple) + (10 * HR) + (2 * RBI) + (2 * R) + (2 * BB) + (2 * HBP) + (5 * SB)



batters['draft_kings_points'] = batters.apply(lambda batter: draft_king_batters(batter['1B'],
                                                                                batter['2B'],
                                                                                batter['3B'],
                                                                                batter['HR'],
                                                                                batter['RBI'],
                                                                                batter['R'],
                                                                                (batter['BB'] + batter['IBB']),
                                                                                batter['HBP'],
                                                                                batter['SB']),axis=1)
```

```python
def draft_king_pitchers(IP,K,W,ER,H,BB,HBP):
    
    tot = (2.25 * IP) + (2 * K) + (-2 * ER) + (-0.6 * H) + (-0.6 * BB) + (-0.6 * HBP)
    
    # Win
    if W == 'W':
        tot +=4
    
    # CG
    if IP == 9:
        tot += 2.5
        
    # CGS
    if IP == 9 and ER == 0:
        tot +=2.5
        
    # No Hitter
    if IP == 9 and H == 0:
        tot += 5
    return tot




pitchers['draft_kings_points'] = pitchers.apply(lambda pitcher: draft_king_pitchers(pitcher['IP'],
                                                                                pitcher['K'],
                                                                                pitcher['Dec'],
                                                                                pitcher['ER'],
                                                                                pitcher['H'],
                                                                                pitcher['BB'],
                                                                                0),axis=1)
```

```python
batters = batters.merge(dkpa, how='left', on='brefID')
batters = batters[batters['DKp/PA'].notna()]
batters = batters[batters['PA'] != 0]
```

```python
batters = batters.sort_values(['date'], ascending=True).reset_index(drop=True)
```

```python
rmse_dict = {}
rmse_summary = {}

def calculate_rmse(batters_df):
    
    dates_list = list(batters_df.date.unique())
    
    for date in dates_list:
        
        #print(date)
        day_df = batters_df.loc[batters_df['date'] == date]
        total_pa = day_df.PA.sum()
        day_df['w'] = day_df.apply(lambda row: row['PA'] / total_pa, axis=1)
        day_df['obs_dkpa'] = day_df.apply(lambda row: row['draft_kings_points'] / row['PA'], axis=1)
        day_df['err'] = day_df.apply(lambda row: row['DKp/PA'] - row['obs_dkpa'], axis=1)
        day_df['sq_err'] = day_df.apply(lambda row: row['err'] ** 2, axis=1)
        day_df['weighted_sq_err'] = day_df.apply(lambda row: row['sq_err'] * row['w'], axis=1)
        
        rmse = sum(day_df['weighted_sq_err'])
        
        rmse_summary[date] = round(rmse, 3)
        
    return rmse_summary
        
    
```

```python
rmse_dict = calculate_rmse(batters)
```

```python
import numpy as np

y_values = []
k = list(rmse_dict.keys())
k.sort()
for i in k: y_values.append(rmse_dict[i])
y = np.array(y_values)

x_values = list(rmse_dict.keys())
#x_values.reverse()
x = np.arange(len(x_values))
```

```python
from numpy.polynomial.polynomial import polyfit
import matplotlib.pyplot as plt

# Fit with polyfit
b, m = polyfit(x, y, 1)

plt.plot(x, y, '.')
plt.plot(x, b + m * x, '-')
plt.show()
```
