```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("2022_batters_jan4.csv")
```

```python
df.head()
```

```python
# points league settings

b1_points = 1
b2_points = 2
b3_points = 3
hr_points = 4
sb_points = 1
r_points = 1
rbi_points = 1
bb_points = 1
so_points = -1
```

```python
def calculate_hitting(b1, b2, b3, hr, sb, r, rbi, bb, so, b1_points = 1, b2_points = 2, b3_points = 3, hr_points = 4, sb_points = 1, r_points = 1, rbi_points = 1, bb_points = 1, so_points = -1):
    return ((int(b1) * b1_points) + 
            (int(b2) * b2_points) + 
            (int(b3) * b3_points) + 
            (int(hr) * hr_points) + 
            (int(sb) * sb_points) +
            (int(r) * r_points) +
            (int(rbi) * rbi_points) +
            (int(bb) * bb_points) +
            (int(so) * so_points))
```

```python
df['FantasyPoints_Hitting'] = df.apply(lambda x: calculate_hitting(x.S, 
                                                                   x.D,
                                                                   x['T'],
                                                                   x.HR,
                                                                   x.SB,
                                                                   x.R,
                                                                   x.RBI,
                                                                   x.BB,
                                                                   x.SO,
                                                                   b1_points,
                                                                   b2_points,
                                                                   b3_points, 
                                                                   hr_points,
                                                                   sb_points, 
                                                                   r_points, 
                                                                   rbi_points, 
                                                                   bb_points, 
                                                                   so_points),axis=1)
```

```python
df['FantasyPoints_Hitting']
```

```python
budget = 260
min_bid = 1
teams = 12
players = 20
avg_budget = budget / players
tot_players = teams * players

print(avg_budget)
print(tot_players)
```

```python
df = df.sort_values(by = 'FantasyPoints_Hitting', ascending= False)

df_top = df.head(tot_players)
df_top
```

```python
minimum = min(df_top['FantasyPoints_Hitting'])

maximum = max(df_top['FantasyPoints_Hitting'])

std = df_top['FantasyPoints_Hitting'].std()

mean = df_top['FantasyPoints_Hitting'].mean()

print(std)
print(mean)
```

```python
def max_std(std, mean, maximum):
    i = 1
    j = 0
    while i > 0:
        if (maximum > mean + (j * std)) & (maximum < mean + (i * std)):
            return i
        else:
            i += 1
            j += 1

sd_above = max_std(std, mean, maximum)
print(sd_above)
```

```python
m = interp1d([minimum,maximum],[min_bid,avg_budget*sd_above])

df_top['auction$'] = round(df_top['FantasyPoints_Hitting'].apply(m), 1)
```
